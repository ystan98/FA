# We can proceed to do fineclassing

We can use this to get a brief idea of the IVs of each variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def woe_iv(data, variable_name, bin_count=10, show_woe=False):

    variable = variable_name
    temp_df = df.copy() 
    temp_df = temp_df[[variable, 'STATUS']]
    temp_df['Bin_Range'] = pd.qcut(df[variable],q=bin_count)

    temp_df['Bin_Range'] = temp_df['Bin_Range'].astype('object')
    temp_df['Bin_Range'].fillna('Missing', inplace = True)

    variable_data = pd.DataFrame()

    variable_data['Bin_Range'] = temp_df.groupby(by='Bin_Range', as_index=False).count()['Bin_Range']


    variable_data['Count'] = temp_df.groupby(by='Bin_Range', as_index=False).count()['STATUS']

    variable_data['Events'] = temp_df.groupby(by='Bin_Range', as_index=False).sum()['STATUS']

    variable_data['Non_Events'] = variable_data['Count'] - variable_data['Events']


    variable_data['%_of_Events'] = variable_data['Events']/sum(variable_data['Events'])

    variable_data['%_of_Non_Events'] = variable_data['Non_Events']/sum(variable_data['Non_Events'])

    variable_data['WOE'] = np.log(variable_data['%_of_Non_Events'] / variable_data['%_of_Events'])

    variable_data['IV'] = (variable_data['%_of_Non_Events']-variable_data['%_of_Events'])*variable_data['WOE']
    
    print(variable_data['WOE'])

    print("For", bin_count, "bins the IV is:")
    IV_ = variable_data['IV'].sum()
    print(np.round(IV_,4))
    
    ffig, ax1 = plt.subplots(figsize=(20,6))
    sns.barplot(x = variable_data['Bin_Range'] , y = variable_data['Count'], alpha=0.5, ax = ax1)
    plt.xticks(rotation=45)

    ax2 = ax1.twinx()

    sns.lineplot(data = variable_data['WOE'], marker='o', sort = False, ax=ax2)

    plt.title("WOE plot for " + variable)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_spare1 = df.copy()
```

```python
for i in df_spare1.columns:
    try:
        woe_iv(df_spare1, i, bin_count=20, show_woe=True)        
    except:
        le = LabelEncoder()
        df_spare1[i] = le.fit_transform(df_spare1[i])
        try:
            woe_iv(df_spare1, i, bin_count=4, show_woe=True)
        except:
            pass# continue in next notebook (coarse classing)
```
