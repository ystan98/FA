```python
corr_matrix = filtered_df.corr().abs()
high_corr_pairs = corr_matrix.stack().sort_values(ascending=False).reset_index()
high_corr_pairs.columns = ['Column 1', 'Column 2', 'Correlation Coefficient']
high_corr_pairs = high_corr_pairs[(high_corr_pairs['Correlation Coefficient'] > 0.8) & (high_corr_pairs['Correlation Coefficient'] <= 0.9) & (high_corr_pairs['Column 1'] != high_corr_pairs['Column 2'])]

# Print the high correlation pairs without same columns
print(high_corr_pairs)
```
